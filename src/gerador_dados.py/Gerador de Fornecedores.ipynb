```python
import pandas as pd
import numpy as np

# 1. Carregar sua tabela de produtos existente
df_produtos = pd.read_csv('dim_produtos.csv')

print("--- Gerando dim_fornecedores baseada nos seus produtos ---")

# 2. Criar os Fornecedores baseados no "Tipo" do produto
# Vamos criar fornecedores fictícios que combinam com os perfis dos seus produtos
dados_fornecedores = [
    {'id_fornecedor': 1, 'nome': 'Indústria MC', 'tipo': 'Marca Própria', 'cidade': 'Alvorada'},
    {'id_fornecedor': 2, 'nome': 'Distribuidora Nacional', 'tipo': 'Líder de Mercado', 'cidade': 'São Paulo'},
    {'id_fornecedor': 3, 'nome': 'Select Importados', 'tipo': 'Premium', 'cidade': 'Porto Alegre'},
    {'id_fornecedor': 4, 'nome': 'Atacado Regional', 'tipo': 'Econômica', 'cidade': 'Canoas'},
    {'id_fornecedor': 5, 'nome': 'Fornecedores Diversos', 'tipo': 'Geral', 'cidade': 'Porto Alegre'}
]

df_fornecedores = pd.DataFrame(dados_fornecedores)
df_fornecedores.to_csv('dim_fornecedores.csv', index=False, encoding='utf-8-sig')
print("Arquivo dim_fornecedores.csv criado!")

# 3. Vincular os produtos aos fornecedores (Engenharia de Atributos)
def definir_fornecedor(nome_produto):
    nome_produto = nome_produto.lower()
    if 'própria' in nome_produto:
        return 1 # Indústria MC
    elif 'líder de mercado' in nome_produto:
        return 2 # Distribuidora Nacional
    elif 'premium' in nome_produto:
        return 3 # Select Importados
    elif 'econômica' in nome_produto:
        return 4 # Atacado Regional
    else:
        return 5 # Diversos

# Aplicar a lógica
df_produtos['id_fornecedor'] = df_produtos['produto'].apply(definir_fornecedor)

# 4. Salvar a dim_produtos atualizada (agora com a coluna id_fornecedor)
df_produtos.to_csv('dim_produtos.csv', index=False, encoding='utf-8-sig')

print("\nSucesso! Sua tabela dim_produtos foi atualizada com o vínculo de fornecedores.")
print(df_produtos[['sku', 'produto', 'id_fornecedor']].head())
```
